```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.pandas.set_option('display.max_columns', None)

df = pd.read_csv('Delhi.csv')
df.head()
```

```python
sns.pairplot(df)
```

```python
plt.scatter(df["Area"],df["Price"])
plt.title("Total Basement area in Square Feet Vs SalePrice ")
plt.ylabel("SalePrice")
plt.xlabel("Total Basement area in Square Feet");
```

```python
sns.factorplot("SwimmingPool","Price",data=df);
```

```python
sns.distplot(df["CarParking"],color='r', kde=False);
```

```python
sns.boxplot(df["PowerBackup"],df["Price"])
plt.title("Sale Price vs Power backup");
```
